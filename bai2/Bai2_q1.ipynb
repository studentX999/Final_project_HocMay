```python
#List of students participated in this report:
#52100896-Pham Gia Huy
#52100844-Nguyen Thi Cam Thuy
#52100243-Nguyen Tuan Kiet
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Đọc dữ liệu từ tập tin CSV
filename = 'car_evaluation.csv'
data = pd.read_csv(filename)
print(data.shape)
print(data.values)

# Hiển thị thông tin cơ bản về dữ liệu
print(data.info())

# Xem một số mẫu dữ liệu
print(data.head())
```

```python
# Thống kê mô tả của dữ liệu
print(data.describe())
```

```python
# Vẽ đồ thị phân phối của các đặc trưng
plt.figure(figsize=(12, 8))
for i, column in enumerate(data.columns[:-1]):
    plt.subplot(2, 3, i + 1)
    sns.countplot(x=column, data=data, hue='decision')
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Encode string features
label_encoder = LabelEncoder()
columns = ['buyingprice','maintenancecost','lug_boot','safety','decision']

for column in data.columns:
    if data[column].dtype == 'object':  # Check if the column contains string data
        data[column] = label_encoder.fit_transform(data[column])

# Vẽ ma trận tương quan
plt.figure(figsize=(10, 8))
sns.heatmap(data.head(30), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
